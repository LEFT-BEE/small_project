```
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

from datetime import datetime 
import pandas_datareader.data as wb
```

```
start = datetime(2012,1,1) 
end = datetime(2021,3,1)

# 주식 종목마다 코드가 다르다. 삼성전자의 코드는 005930
samsung = wb.DataReader('005930.KS','yahoo',start,end) 
kakao = wb.DataReader('035720.KS' , 'yahoo' , start,end)
naver = wb.DataReader('035420.KS' , 'yahoo' , start,end)
```

```
date_list = ["2012-07-16" , "2013-07-16" , "2014-07-16",
"2015-07-16" , "2016-07-20", "2017-07-20", "2018-07-16","2019-07-16","2020-03-20",
"2021-03-02"]
x_list = ['2012' , '2013' , '2014' , '2015' , '2016' , '2017' , '2018' , '2019' , '2020' , '2021']
bank_ =[3.0 , 2.5 , 2.25 , 1.5 , 1.25 , 1.5 , 1.75 , 1.5 , 0.5 , 0.5]
```

```
ss=[]
kao = []
nvr =[]
for date in date_list:
  ss.append(samsung["Open"][date])
  kao.append(kakao["Open"][date])
  nvr.append(naver["Open"][date])

```

```
fig , ax = plt.subplots()
ax2 = ax.twinx()
ax.plot(x_list , kao ,'y--' , marker="o", label ="Kakao")
ax.plot(x_list , nvr , 'g-' ,marker = 's',label = "Naver")
ax.set_xlabel('year')
ax.set_ylabel("stock price (won)")
ax.legend(bbox_to_anchor=(1.4,1))

ax2.plot(x_list , bank_ ,':' ,marker = "*", label = "Interest Rate")
ax2.set_ylabel("the fluctuation of interest rate")
ax2.legend(bbox_to_anchor=(1.5,0.8))

plt.title("Stocks by Year & fluctuation of interest rate")
plt.show()
```
